```python
# Import Libraries

import pandas as pd
import numpy as np
import re
```

```python
# get the data from my activity

my_activity = pd.read_csv('../data/my_activity_clean.csv', sep=';', decimal=',', index_col=0)
```

```python
my_activity.head()
```

```python
# Get only data from youtube

youtube = my_activity[my_activity['products'] == 'YouTube'].reset_index(drop=True)
youtube.head()
```

```python
youtube.columns
```

```python
youtube = youtube[['title', 'titleUrl', 'time']]
```

```python
def matching(x):
    if re.match(r'(^Searched|^Watched)(?:\sfor)?\s(.*)',x):
        return False
    else:
        return True
```

```python
youtube = youtube.drop(youtube[youtube['title'].apply(matching)].index[0]).reset_index(drop=True)
```

```python
def get_searches_watches(string,numgroup):
    result = re.search(r'(^Searched|^Watched)(?:\sfor)?\s(.*)',string)
    return result.group(numgroup)
    

get_searches_watches('Searched for trailers 2014 movies official',1)
```

```python
youtube['type'] = youtube['title'].apply(lambda x: get_searches_watches(x,1))
youtube['content'] = youtube['title'].apply(lambda x: get_searches_watches(x,2))
```

```python
youtube = youtube.drop('title', axis = 1)
```

```python
youtube
```

```python
youtube.to_csv('../data/youtube.csv', sep=';', decimal=',')
```

```python
# create the count of days of searches and watches by day

youtube_by_day = youtube
```

```python
youtube_by_day['date'] = youtube_by_day['time'].apply(lambda x: re.findall(r'\d{4}-\d{2}-\d{2}',x)[0])
```

```python
youtube_by_day = youtube_by_day.groupby(['date', 'type'])['type'].count()

```

```python
youtube_by_day.name = 'count'
youtube_by_day = youtube_by_day.reset_index()
```

```python
youtube_by_day = youtube_by_day.pivot(index = 'date', columns = 'type', values = 'count').reset_index()
```
