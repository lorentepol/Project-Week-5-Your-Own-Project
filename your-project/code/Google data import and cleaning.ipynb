```python
import os
import pandas as pd
import numpy as np
import json
import dateutil
```

```python
rootdir_google_activity= r"C:\Users\Pol\Documents\Ironhack\Projects\Project-Week-5-Your-Own-Project\your-project\data\Google\My Activity"
```

```python
my_activity_df = pd.DataFrame()
for path,dirs,files in os.walk(rootdir_google_activity):
    for name in files:
        data_extracted = pd.read_json(path + r'\\' + name)
        my_activity_df = pd.concat([my_activity_df, data_extracted], ignore_index=True)
```

```python
my_activity_df = my_activity_df.sort_values('time').reset_index(drop=True)

my_activity_df
```
